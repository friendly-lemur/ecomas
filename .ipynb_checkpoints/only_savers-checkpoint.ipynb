# Market Model - Notebook

### Building Model

```python
import mesa
from mesa.time import RandomActivation
from mesa.time import RandomActivationByType
from mesa.datacollection import DataCollector
```

```python
class Agent_Saver(mesa.Agent):
    def __init__(self, unique_id, model,
                 position,
                 wealth, order, waiting, invested):
        super().__init__(unique_id, model)
        
        # DOF Saver
        self.position = position # initial position, buy or sell
        
        self.wealth = wealth # cash
        self.order = order # quantities ordered
        self.waiting = waiting # timesteps of waiting before buying
        self.invested = invested
        
    # Saver Functions
    # the savers keep the position since the beginning of simulation
    # and keep buying or selling during the simulation
    def invest(self):
        if self.position == 'buy':
            self.order = 1 # choose a quantity to buy
            # set self.waiting = qualcosa prima di comprare di nuovo
        elif self.position == 'sell':
            self.order = 1 # choose a quantity to sell
        
        # update remaining wealth
        self.wealth = self.wealth - self.order * model.Price
    
    # Step Function for saver agent
    # if they have no more wealth they're out of money to invest
    # assume savers have a constant outside income they invest periodically
    def step(self):
        self.wealth += 5 # income of a saver
        
        if self.wealth <= 0:
            self.position = 'pass'
        
        if self.invested == True:
            self.order = 0 # do not buy anymore but wait
            self.waiting -= 1
            if self.waiting == 0:
                self.invested = False
        else:
            self.invest()
            self.invested = True
            self.waiting = 2
```

```python
# N_agents = 10,100,..., 1000 and so on
# p_savers = percentage of savers among the N_agents
# savers_sentiment = {0,1}, e.g. 0.60 => 60% savers are buyers and 40% sellers
# Price = starting price of the asset, e.g. 12

class Market_Model(mesa.Model):
    """Market Model with buyers and sellers"""

    def __init__(self, N_agents, Price, p_savers, savers_sentiment):
        
        # DOF model
        self.N_agents = N_agents
        self.Price = Price
        
        # Environment Settings
        self.schedule = mesa.time.RandomActivation(self)
        self.savers_sentiment = savers_sentiment
        
        # Create savers
        N_savers = int(N_agents * p_savers)
        wealth = 1000
        order = 0
        waiting = 3
        invested = True
        
        print('Creating saver agents...')
        for i in range(N_savers):
            print(i)
            if i < N_savers * savers_sentiment:
                position = 'buy'
            else:
                position = 'sell'
            
            a = Agent_Saver(i, self, position, wealth, order, waiting, invested)
            self.schedule.add(a)
        
        print('Creating random traders.')
        # N_rnd_traders = int(N_agents * p_rnd_traders)
        N_rnd_traders = N_agents - N_savers
        
        
        # Create agent....
        #for i in range(self.num_buyers):
        #    a = Agent(i, self)
        #    self.schedule.add(a)
               
        self.datacollector = DataCollector(
            model_reporters = {'Price': change_Prices}, # here put methods for datacollecting
            agent_reporters = {
                'Wealth': 'wealth',
                'Position': 'position',
                'Order': 'order'
            }
        )
    
    def step(self):
        """Advance the model by one step."""
        self.schedule.step()
        
        # Collect the data at that timestep
        self.datacollector.collect(self)
```

```python
def change_Prices(model):
    agent_wealth = [agent.wealth for agent in model.schedule.agents]
    agent_positions = [agent.position for agent in model.schedule.agents]
    agent_orders = [agent.order for agent in model.schedule.agents]
    
    sbo = 0
    sso = 0
    alfa = 0.1
    
    for i in range(len(agent_positions)):
        if agent_positions[i] == 'buy':
            sbo += agent_orders[i]
        elif agent_positions[i] == 'sell':
            sso += agent_orders[i]
    
    delta = alfa * (sbo - sso)
    model.Price = model.Price + delta
    
    return model.Price
```

### Running Model

```python
N_agents = 10
Price = 10
p_savers = 1.0
savers_sentiment = 0.60

model = Market_Model(N_agents, Price, p_savers, savers_sentiment)

for i in range(20):
    model.step()
```
