## 3.8 Augmented Dickey-Fuller Test

```python
from statsmodels.tsa.stattools import adfuller
```

A time series is said to be “stationary” if it has no trend, exhibits constant variance over time, and has a constant autocorrelation structure over time.  
One way to test whether a time series is stationary is to perform an augmented Dickey-Fuller test, which uses the following null and alternative hypotheses:  
- H0: The time series is non-stationary. In other words, it has some time-dependent structure and does not have constant variance over time.  
- HA: The time series is stationary.

So:
- p-value > 0.05: Fail to reject the null hypothesis (H0), the data has a unit root and is non-stationary.
- p-value <= 0.05: Reject the null hypothesis (H0), the data does not have a unit root and is stationary.


```python
_ = adfuller(prices_df.true)
```

```python
print('Null Hypothesis: The time series is non-stationary (there IS a unit root).')
print(f'Test Statistic: {_[0]:.2f}')
print(f'Test p-value: {_[1]:.2f}')
print(f'Chosen significance level: {0.05}')
```

```python
_ = adfuller(prices_df.pred.dropna())
```

```python
print('Null Hypothesis: The time series is non-stationary (there IS a unit root).')
print(f'Test Statistic: {_[0]:.2f}')
print(f'Test p-value: {_[1]:.2f}')
print(f'Chosen significance level: {0.05}')
```

## 3.9 Volatility Clustering
Qualitatively, it means that high returns are likely to be followed by high returns, the same applying for low returns.  
Quantitatively, it means that the series of absolute returns will exhibit a significant and slowly decaying pattern as in the plot you showed above.  

```python
# Compute log returns and absolute returns
log_returns = np.log(prices_df['true']) - np.log(prices_df['true'].shift(1))
absolute_returns = np.abs(log_returns)

# Compute autocorrelation functions for log returns and absolute returns
log_returns_acf = [log_returns.autocorr(lag=lag) for lag in range(21)]
absolute_returns_acf = [absolute_returns.autocorr(lag=lag) for lag in range(21)]

# Plot autocorrelation functions
lags = range(21)
fig, ax = plt.subplots(2, 1, figsize=(8, 8))
ax[0].stem(lags, log_returns_acf)
ax[0].set(title='Autocorrelation Function of Log Returns - Real Market', xlabel='Lag', ylabel='ACF')
ax[1].stem(lags, absolute_returns_acf)
ax[1].set(title='Autocorrelation Function of Absolute Returns - Real Market', xlabel='Lag', ylabel='ACF')
plt.tight_layout()
plt.show()
```

```python
# Compute log returns and absolute returns
log_returns = np.log(prices_df['pred']) - np.log(prices_df['pred'].shift(1))
absolute_returns = np.abs(log_returns)

# Compute autocorrelation functions for log returns and absolute returns
log_returns_acf = [log_returns.autocorr(lag=lag) for lag in range(21)]
absolute_returns_acf = [absolute_returns.autocorr(lag=lag) for lag in range(21)]

# Plot autocorrelation functions
lags = range(21)
fig, ax = plt.subplots(2, 1, figsize=(8, 8))
ax[0].stem(lags, log_returns_acf)
ax[0].set(title='Autocorrelation Function of Log Returns - Artificial Market', xlabel='Lag', ylabel='ACF')
ax[1].stem(lags, absolute_returns_acf)
ax[1].set(title='Autocorrelation Function of Absolute Returns - Artificial Market', xlabel='Lag', ylabel='ACF')
plt.tight_layout()
plt.show()
```

The ACF measures the correlation between a series and its lagged values, and it can take on positive or negative values depending on whether the series tends to be positively or negatively correlated with itself at different lags.  

In the case of absolute returns, which measure the magnitude of the price changes regardless of their direction, the sign of the returns is not taken into account. As a result, the ACF of the absolute returns can take on negative values if there is a tendency for large price changes to be followed by smaller ones, or vice versa. This is known as volatility clustering, and it is a common phenomenon in financial time series.  

## 3.10 Fat Tail Phenomenon
Study the distribution of returns and if it's leptokurtic than we have the fat tail phenomenon (distribution is more outliner-prone than the normal distribution). Also here consider the returns and absolute returns.

```python
from scipy import stats
```

```python
# Returns
print(r_ret.shape) # real market returns
print(s_ret.shape) # simulated market returns
```

```python
def compare_plot(r_ret, s_ret, start, end):
    fig, ax = plt.subplots(ncols = 1, figsize = (7,4))
    fig.suptitle(f'Real vs Simulated returns (from {start} to {end})')
    
    a = 0.40 # binrange
    bw = 0.02 # binwidth
    
    sns.histplot(
        data = r_ret,
        element = 'bars', fill = True, stat = 'density',
        binwidth = bw, binrange = (-a, +a),
        ax = ax, label = 'real'
    )

    sns.histplot(
        data = s_ret,
        element = 'bars', fill = True, stat = 'density',
        binwidth = bw, binrange = (-a,+a),
        ax = ax, label = 'simulated'
    )
    
    ax.set_xlabel('Returns')
    ax.set_ylabel('Density')
    ax.legend()
    
    plt.show();
```

```python
compare_plot(r_ret, s_ret, start, end)
```

```python
r_mean = np.mean(r_ret[1:])
r_variance = np.var(r_ret[1:])
r_kurtosis = stats.kurtosis(r_ret[1:])

s_mean = np.mean(s_ret)
s_variance = np.var(s_ret)
s_kurtosis = stats.kurtosis(s_ret)

print(f'Mean real:    \t {round(r_mean, 4)}  \tMean simulated: \t{round(s_mean, 4)}')
print(f'Variance real:\t {round(r_variance, 4)}  \tVariance simulated: \t{round(s_variance, 4)}')
print(f'Kurtosis real:\t {round(r_kurtosis, 4)}  \tKurtosis simulated: \t{round(s_kurtosis, 4)}')
```

We look at the kurtosis and since for the real market it is 5, then we can consider it leptokurtic (> 3).

### One sample Z-test on Kurtosis
We compare the single value of the kurtosis obtained from the real market with a collection of kurtosis values which come from independent iterations of the same model, changing the random seed.

```python
index = [0, 17, 20, 23, 27] # indices that never converge
random_seeds = np.delete(np.arange(0, 60, 1), index)
```

```python
print(random_seeds)
```

```python
%%time
N_steps = 730

array_Price_values = np.zeros(N_steps)

for rnd_state in random_seeds:
    random.seed(int(rnd_state))
    print(rnd_state)
    # re-istantiate the model
    model = Market_Model_4(N_agents, Price, p_savers, p_followers, p_whales, savers_sentiment, WLT, p_passing_followers, p_passing_whales)
    
    for i in range(N_steps):
        print(i)
        model.step()

    array_Price_values_new = model.datacollector.get_model_vars_dataframe().values # numpy array
    array_Price_values = np.vstack((array_Price_values, np.hstack(array_Price_values_new)))
```

```python
prices_samples = array_Price_values[1:] # removing the init array (full of zeros)
print(prices_samples.shape)
```

```python
df_prices_samples = pd.DataFrame(data = prices_samples.T, columns = random_seeds.tolist())
display(df_prices_samples)
```

```python
# Now transform them into returns
returns_samples = np.log(df_prices_samples / df_prices_samples.shift(1)).dropna()
```

```python
print(returns_samples.shape)
```

```python
s_kurtosis = list()
for it in random_seeds:
    s_kurtosis.append(stats.kurtosis(returns_samples[it]))
```

```python
from statsmodels.stats.weightstats import ztest as ztest
_ = ztest(s_kurtosis, value = r_kurtosis) # compare the list of kurtosis with the true value of real market kurtosis
```

```python
print('Null Hypothesis: The simulated returns have the same kurtosis as the real returns')
print(f'Test Statistic: {_[0]:.2f}')
print(f'Test p-value: {_[1]:.2f}')
print(f'Chosen significance level: {0.05}')
```

When p-value is > 0.05, we do not have sufficient evidence to reject the null hypothesis.  
When p-value is <= 0.05, we do have sufficient evidence to reject the null hypothesis.
